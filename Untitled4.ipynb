VADER:

```python
#bag of words approach using VADER sentiment analysis tool
from nltk.sentiment import SentimentIntensityAnalyzer

#tqdm used for progression bar
from tqdm.notebook import tqdm

sia=SentimentIntensityAnalyzer()
```

```python
#getting the sentiment scores dictionnary 
sia.polarity_scores("I am so happy!")
#compound is the intensity of the sentiment.
```

```python
sia.polarity_scores("this is the worst thing ever.")
```

```python
sia.polarity_scores(random_review)
```

```python
#get the polarity score of each review on the dataset
res={}
for i,row in tqdm(df.iterrows(),total=len(df)):
    text=row['Text']
    myid=row["Id"]
    res[myid]=sia.polarity_scores(text)
    
    
```

```python
#converting the dictionnary of scores to a pandas dataframe
vaders=pd.DataFrame(res).T
#merging the scores dataframe with the initial dataframe
vaders=vaders.reset_index().rename(columns={'index':'Id'})
vaders=vaders.merge(df,how="left")
#new dataframe contains metadata and scores
vaders
```

```python
#plotting VADER results
ax=sns.barplot(data=vaders,x="Score",y="compound")
ax.set_title("compound score by amazon star review")
```

```python
fig,axes=plt.subplots(1,3,figsize=(15,5))
ax=sns.barplot(data=vaders,x="Score",y="pos",ax=axes[0])
ax=sns.barplot(data=vaders,x="Score",y="neu",ax=axes[1])
ax=sns.barplot(data=vaders,x="Score",y="neg",ax=axes[2])
axes[0].set_title("positive")
axes[1].set_title("neutral")
axes[2].set_title("negative")

```

```python
sfrom transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
# Defining the model and tokenizer to use, in this case we use "cardiffnlp/twitter-roberta-base-sentiment" model 
# a model fine-tuned specifically for sentiment analysis
MODEL=f"cardiffnlp/twitter-roberta-base-sentiment"
#creating an instance of the tokenizer associated with the specified pre-trained model
#The tokenizer is used to convert input text into tokenized format that can be processed by the model.
tokenizer=AutoTokenizer.from_pretrained(MODEL)
#creating an instance of the model associated with the specified pre-trained model.
#This model is pre-trained on a large corpus of text and fine-tuned for the specific sentiment classification task.
model=AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
print(random_review)
sia.polarity_scores(random_review)
```

```python
#run for the roberta model
encoded_text=tokenizer(random_review,return_tensors='pt')
output=model(**encoded_text)

scores=output[0][0].detach().numpy()
#The detach() method is used to detach the tensor from the computation graph
#numpy() converts it to a NumPy array.

#applying the softmax function to the logits. to obtain probability scores for different sentiment classes.
scores=softmax(scores)

scores_dict={"roberta_neg":scores[0],"roberta_neu":scores[1],"roberta_pos":scores[2]}
print(scores_dict)
```

```python
def polarity_scores_roberta(random_review):
    encoded_text=tokenizer(random_review,return_tensors='pt')
    output=model(**encoded_text)
    scores=output[0][0].detach().numpy()
    scores=softmax(scores)
    scores_dict={"roberta_neg":scores[0],"roberta_neu":scores[1],"roberta_pos":scores[2]}
    return scores_dict
```

```python
#get the polarity score of each review on the dataset
res={}
for i,row in tqdm(df.iterrows(),total=len(df)):
    try:
        text=row['Text']
        myid=row["Id"]
        vader_res=sia.polarity_scores(text)
        vader_res_rename={}
        for k,v in vader_res.items():
            vader_res_rename[f"vader_{k}"]=v
        roberta_result=polarity_scores_roberta(text)
        both_results={**vader_res_rename,**roberta_result}
        res[myid]=both_results
    except RuntimeError:
        print(f'broke for id {myid}')
    
    
```

```python
result_df=pd.DataFrame(res).T
result_df=result_df.reset_index().rename(columns={'index':'Id'})
result_df=result_df.merge(df,how="left")
```
